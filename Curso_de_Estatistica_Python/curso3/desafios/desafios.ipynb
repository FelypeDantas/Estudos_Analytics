## Aula 2: Análise exploratória dos dados

Para os desafios dessa aula, vamos utilizar a mesma base de dados que estamos utilizando para o desenvolvimento do nosso projeto. Sendo assim, na célula abaixo vamos realizar a importação dessa base de dados e aplicar as transformações já realizadas até o momento da aula 2.

```python
import pandas as pd
```

```python
# importando a base de dados
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
                      'Prédio Inteiro', 'Loja/Salão',
                      'Galpão/Depósito/Armazém',
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']

# removendo imóveis comerciais
df = dados.query('@imoveis_comerciais not in Tipo')

# selecionando apenas imóveis do tipo apartamento
df = df.query('Tipo == "Apartamento"')

# visualizando a base de dados
df.head()
```

### Desafio 1
Calcular a média de quartos por apartamento

```python
dados['Quartos'].mean()
```

### Desafio 2
Conferir quantos bairros únicos existem na nossa base de dados

```python
dados['Bairro'].unique()
```

```python
dados['Bairro'].nunique()
```

### Desafio 3
Analisar quais bairros possuem a média de valor de aluguel mais elevadas

```python
# Criar o DataFrame
df = pd.DataFrame(dados)

# Agrupar por bairro e calcular a média do valor do aluguel
media_aluguel = df.groupby('Bairro')['Valor'].mean().reset_index()

# Ordenar por valor médio de aluguel (decrescente)
media_aluguel = media_aluguel.sort_values(by='Valor', ascending=False)

# Exibir os bairros com média de aluguel mais elevada
print(media_aluguel)
```

### Desafio 4
Criar um gráfico de barras verticais que apresente os 5 bairros com a média de valor de aluguel mais elevadas

```python
import matplotlib.pyplot as plt
```

```python
# Selecionar os 5 bairros com as maiores médias
top_5_bairros = media_aluguel.head(5)\

# Criar o gráfico de barras verticais
plt.figure(figsize=(10, 6))
plt.bar(top_5_bairros['Bairro'], top_5_bairros['Valor'], color='skyblue', edgecolor='black')
```

```python
# Adicionar títulos e rótulos
plt.title('Top 5 Bairros com Maior Média de Aluguel', fontsize=16)
plt.xlabel('Bairro', fontsize=12)
plt.ylabel('Média do Valor de Aluguel (R$)', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```

### Desafio 1
Verifique se a base de dados possui dados nulos e caso tenha, realize o tratamento desses dados nulos

```python
df = pd.DataFrame(dados)
```

```python
# Verificar valores nulos antes do tratamento
print("Valores nulos antes do tratamento:")
print(df.isnull().sum())
```

```python
# Tratamento de valores nulos
for col in df.columns:
    if df[col].dtype in ['float64', 'int64']:  # Colunas numéricas
        df[col].fillna(0, inplace=True)
    elif df[col].dtype == 'object':  # Colunas de texto
        df[col].fillna('', inplace=True)

# Verificar valores nulos após o tratamento
print("\nValores nulos após o tratamento:")
print(df.isnull().sum())

# Exibir DataFrame atualizado
print("\nDataFrame tratado:")
print(df)
```

```python
# Importando novamente os dados
dados_1 = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv')
```

### Desafio 1
Os alunos participaram de uma atividade extracurricular e ganharam pontos extras. Esses pontos extras correspondem a 40% da nota atual de cada um deles. Com base nisso, crie uma coluna chamada "Pontos_extras" que contenha os pontos extras de cada aluno, ou seja, 40% da nota atual deles

```python
dados_1['Pontos_extras'] = dados_1['Notas'].apply(lambda x: x*0.4)
dados
```

### Desafio 2
Crie mais uma coluna, chamada "Notas_finais" que possua as notas de cada aluno somada com os pontos extras

```python
dados_1['Notas_finais'] = dados_1['Notas'] + dados_1['Pontos_extras']
dados_1
```

### Desafio 3
Como os alunos ganharam pontos extras, alguns que não tinham sido aprovados antes podem ter sido aprovados agora. Com base nisso, crie uma coluna chamada "Aprovado_final" com os seguintes valores:

* True: caso o aluno esteja aprovado (nota final deve ser maior ou igual a 6);
* False: caso o aluno esteja reprovado (nota final deve ser menor que 6).


```python
dados_1['Aprovado_final'] = dados_1['Notas_finais'].apply(lambda x: True if x >= 6 else False)
dados_1
```

### Desafio 4
Faça uma seleção e verifique quais alunos não tinham sido aprovados anteriormente, mas foram aprovados após a soma dos pontos extras

```python
dados_1.query('Aprovado == False & Aprovado_final == True')
```

```python
selecao = (dados_1['Aprovado'] == False) & (dados_1['Aprovado_final'] == True)
dados_1[selecao]
```
